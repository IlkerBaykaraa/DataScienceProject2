First, let's start with the graph requested in the second question. A graph was created using two properties of the elements in the iris dataset (sepal_length and sepal_width). As can be seen in this chart, 3 species separated according to the "species" feature are shown. Then, this graph was divided into the desired number of clusters with the K-Means Cluster method. The K-Means Cluster method performed k clustering operations according to the entered k value. Clustering was done in such a way that the similarity between the cluster elements was max and the similarity between clusters was min. After entering the k value, k points were created as the first step. By comparing the distances of these points, called the center point, to other points, the closest point was included in the set of that center. New centers were obtained by taking the midpoint of the clusters formed, and this situation continued until there was no change. As mentioned in the first part, the aim was to maximize the similarity between the cluster elements. Thanks to this method, the elements separated according to the types in the first graph were separated according to the similarity ratio among themselves. As a result, the K-Means Cluster method allowed the data to be analyzed more easily.
